```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sentence = "What are the symptoms of diabetes?"
tokens = sentence.lower().replace('?', '').split()
```

```python
np.random.seed(42)          # Reproducibility
d_model = 64                # Embedding dimension
d_k = 64                    # Key/Query/Value dimension
n = len(tokens)             # Sequence length
```

```python
X = np.random.randn(n, d_model)   # Shape: (6, 64)
```

```python
W_Q = np.random.randn(d_model, d_k)   # Shape: (64, 64)
W_K = np.random.randn(d_model, d_k)   # Shape: (64, 64)
W_V = np.random.randn(d_model, d_k)   # Shape: (64, 64)
```

```python
Q = X @ W_Q    # Queries:  shape (n, d_k)
K = X @ W_K    # Keys:     shape (n, d_k)
V = X @ W_V    # Values:   shape (n, d_k)
```

```python
scores = Q @ K.T / np.sqrt(d_k)   # Shape: (n, n)
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x, axis=-1, keepdims=True))  # Stability
    return e_x / e_x.sum(axis=-1, keepdims=True)

attention_weights = softmax(scores)   # Shape: (n, n)
```

```python
attention_output = attention_weights @ V   # Shape: (n, d_k)
```

```python
print("Tokens:", tokens)
print(f"\nInput Embedding Shape: {X.shape}")
print(f"Raw Attention Score Matrix Shape: {scores.shape}")
print(f"Attention Weight Matrix Shape:    {attention_weights.shape}")
print(f"Attention Output Shape:           {attention_output.shape}")

print("\n── Attention Weight Matrix (rows = query, cols = key) ──")
print("Tokens:", tokens)
for i, row in enumerate(attention_weights):
    formatted = "  ".join(f"{v:.4f}" for v in row)
    print(f"  [{tokens[i]:10s}]  {formatted}")
```
